```python
import numpy as np
import pandas as pd
```

```python
filename = "MeToo_093017_101518_2.csv"
tweets_df = pd.read_csv(filename)
n = tweets_df.shape[0]  # data numbers
```

```python
# top retweets
top_n = max(int(n*0.0005), 10)  # top 0.05 percent or 10 tweets
topretweets_df = tweets_df.nlargest(top_n, 'retweets')
topretweets_df[['retweets','favorites','text']]
```

```python
# top favorites
topfavorites_df = tweets_df.nlargest(top_n, 'favorites')
topfavorites_df[['retweets','favorites','text']]
```

```python
# top retweets + favorites
tweets_df['sums'] = tweets_df['retweets'] + tweets_df['favorites']
topsums_df = tweets_df.nlargest(top_n, 'sums')
topsums_df[['sums','retweets','favorites','text']]
```

```python
# top retweets to favorites ratio 
filtTweets_df = tweets_df[tweets_df.retweets > 100] # only analyze popular tweets 
assert(all(filtTweets_df['favorites'] > 0))
ratio = filtTweets_df.retweets/filtTweets_df.favorites
filtTweets_df.loc[:,'ratios'] = ratio
```

```python

topratios_df = filtTweets_df.nlargest(top_n, 'ratios')
topratios_df[['ratios','retweets','favorites','text']]
```

```python
# top favorites to retweets ratio
invratios_df = filtTweets_df.nsmallest(top_n, 'ratios')
invratios_df[['ratios','retweets','favorites','text']]
```
